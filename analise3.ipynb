# As perguntas podem ter resposta diretas(onde a visualização dos dados apresenta a solução) ou criar novas colunas na tabela a partir de processamento dos dados ex:

```python
df['pg'] = 1 #criando uma nova coluna e colocando '1', ou seja, que todas tiveram pagamento
df
```

```python
df.loc[df.dt_pgto.isnull(), 'pg'] = 0 # se nãao tem data de pagamento é porque não tem pagamento, criando mais uma coluna e processando
df

```

```python
contagem_pg = df['pg'].value_counts()

# Extraindo os valores
pagos = contagem_pg.get(1, 0)  # Retorna 0 se não houver o valor 1
nao_pagos = contagem_pg.get(0, 0)  # Retorna 0 se não houver o valor 0           

print("Quantidade de registros pagos: ", pagos)
print("Quantidade de registros não pagos: ", nao_pagos)


#extraído do chatGPT
```

```python
df['tempo_pg'] = df.dt_pgto - df.dt_venda #DESCOBRINDO o tempo em dias entre a venda e pagamento e criando uma coluna
df
```
